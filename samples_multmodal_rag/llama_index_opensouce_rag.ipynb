```python
import os
import shutil
import PyPDF2
from pdf2image import convert_from_path
from docx import Document
import docx2txt
from pathlib import Path

# Function to extract text from PDF files
def extract_pdf_text(file_path, output_folder):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()

        output_file = os.path.join(output_folder, Path(file_path).stem + '.txt')
        with open(output_file, 'w', encoding='utf-8') as output:
            output.write(text)

# Function to extract images from PDF files
def extract_pdf_images(file_path, output_folder):
    pages = convert_from_path(file_path)
    for i, page in enumerate(pages):
        image_path = os.path.join(output_folder, f'{Path(file_path).stem}_image_{i+1}.png')
        page.save(image_path, 'PNG')

# Function to extract text from DOCX files
def extract_docx_text(file_path, output_folder):
    doc = Document(file_path)
    text = '\n'.join([para.text for para in doc.paragraphs])

    output_file = os.path.join(output_folder, Path(file_path).stem + '.txt')
    with open(output_file, 'w', encoding='utf-8') as output:
        output.write(text)

# Function to extract images from DOCX files
def extract_docx_images(file_path, output_folder):
    # Extract images directly to the output folder without subfolder
    docx2txt.process(file_path, output_folder)

# Function to process all files in a folder
def process_files_from_folder(folder_path):
    output_folder = "output1"
    if os.path.exists(output_folder):
        shutil.rmtree(output_folder)  # Clear previous output
    os.makedirs(output_folder)  # Create a fresh output folder

    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)

        if file_name.endswith('.pdf') or file_name.endswith('.docx'):
            print(f"Processing: {file_path}")

            if file_name.endswith('.pdf'):
                extract_pdf_text(file_path, output_folder)
                extract_pdf_images(file_path, output_folder)
            elif file_name.endswith('.docx'):
                extract_docx_text(file_path, output_folder)
                extract_docx_images(file_path, output_folder)
        else:
            print(f"Unsupported file type: {file_path}")

# Main function
if __name__ == "__main__":
    folder_path = "/content"  # Specify the folder path here
    process_files_from_folder(folder_path)
    print("Processing complete. Check the output folder for results.")
```

```python
import os
from llama_index.core import SimpleDirectoryReader #Updated import statement
from llama_index.core import StorageContext, ServiceContext
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.embeddings import OpenAIEmbedding
from llama_index.core.indices import MultiModalVectorStoreIndex
from llama_index.llms import OpenAI
import chromadb
from dotenv import load_dotenv
from llama_index.multi_modal_llms import OpenAIMultiModal

# Load environment variables (make sure to set OPENAI_API_KEY in your .env file)
load_dotenv()

# Specify a persistent directory for ChromaDB
persist_directory = "./chroma_db1"
client = chromadb.PersistentClient(path=persist_directory)

# Create text and image vector stores using ChromaDB
text_collection = client.get_or_create_collection("text_collection")
image_collection = client.get_or_create_collection("image_collection")

text_store = ChromaVectorStore(chroma_collection=text_collection)
image_store = ChromaVectorStore(chroma_collection=image_collection)

# Create storage context using the text and image vector stores
storage_context = StorageContext.from_defaults(
    vector_store=text_store,
    image_store=image_store
)

# Set up OpenAI embeddings and GPT-4 Vision model
embed_model = OpenAIEmbedding()
llm = OpenAI(model="gpt-4")
multi_modal_llm = OpenAIMultiModal(model="gpt-4-vision-preview", max_new_tokens=300)

# Create a service context with the specified models
service_context = ServiceContext.from_defaults(
    llm=llm,
    embed_model=embed_model,
    multi_modal_llm=multi_modal_llm
)

# Load documents from directory
documents = SimpleDirectoryReader("/content/output1").load_data()

# Create the MultiModal index using the documents, storage context, and service context
index = MultiModalVectorStoreIndex.from_documents(
    documents,
    storage_context=storage_context,
    service_context=service_context,
)

print("Multimodal index built successfully with persistent storage!")

# Example of querying the index
query_engine = index.as_query_engine()
response = query_engine.query("Describe the images in the dataset.")
print(response)
```

```python
from llama_index.core.response.notebook_utils import display_source_node
from llama_index.core.schema import ImageNode
import os
from PIL import Image
import matplotlib.pyplot as plt

MAX_TOKENS = 1049
retriever_engine = index.as_retriever(
    similarity_top_k=5, image_similarity_top_k=1
)

def retrieve_and_display(query):
    retrieval_results = retriever_engine.retrieve(query[:MAX_TOKENS])

    retrieved_images = []
    for res_node in retrieval_results:
        if isinstance(res_node.node, ImageNode):
            retrieved_images.append(res_node.node.metadata["file_path"])

        else:
            display_source_node(res_node, source_length=200)

    if retrieved_images:
        plt.figure(figsize=(15, 5))
        for i, img_path in enumerate(retrieved_images):
            plt.subplot(1, len(retrieved_images), i+1)
            img = Image.open(img_path)
            plt.imshow(img)
            plt.axis('off')
        plt.show()

# Example usage
retrieve_and_display("saint joseph bedroom")
```
